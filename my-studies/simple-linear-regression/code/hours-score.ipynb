## LINEAR REGRESSION MODEL

```python
#Dependent and Independent Variable
X = dataset.iloc[:,0:-1].values
Y = dataset.iloc[:,-1].values
```

```python
print(X)
```

```python
print(Y)
```

```python
#Split the dataset into train and test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=123)
```

```python
print(X_train, X_test)
```

```python
print(Y_train, Y_test)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, Y_train)
```

```python
reg.score(X_train, Y_train)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
y_pred = reg.predict(X_test)
```

```python
print(y_pred)
```

```python
scatter = sns.scatterplot(x = np.ravel(X_train), y = np.ravel(Y_train))
sns.lineplot(np.ravel(X_train), reg.predict(X_train), color = 'red')
scatter.set(ylabel='Score', xlabel='Hours', title='Study Hours x Score - Training Set')
plt.show()
```

```python
scatter = sns.scatterplot(x = np.ravel(X_test), y = np.ravel(Y_test))
sns.lineplot(np.ravel(X_train), reg.predict(X_train), color = 'red')
scatter.set(ylabel='Score', xlabel='Hours', title='Study Hours x Score - Test Set')
plt.show()
```
